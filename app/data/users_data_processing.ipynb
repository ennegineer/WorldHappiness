```python
# Dependencies, files etc.
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.ticker import FuncFormatter
import numpy as np

netfile = 'raw/InternetUsers.csv'

# Read file and store in dataframe
userdata = pd.read_csv(netfile)
```

```python
# Inspect the internet users data.
userdata.head()
```

```python
# Drop data for years prior to 2005, since the World Happiness data starts with 2005.
users_present = userdata[userdata['Year'] >= 2005]
users_present.head(15)
```

```python
# Create a dataframe for the users by country, removing any containing 'income'
indexincomes = users_present[users_present['Entity'].str.contains('income') == True].index
users_bycountry = users_present.drop(indexincomes)
print(users_bycountry)
```

```python
users_bycountry.reset_index()
users_bycountry.drop('Unnamed: 0', axis = 1)
```

```python
# Drop the regions from the country dataframe
indexregions = users_bycountry[users_bycountry['Code'] == 'Region'].index
final_users = users_bycountry.drop(indexregions)
print(final_users)
```

```python
final_users.reset_index()
final_users.drop('Unnamed: 0', axis = 1)
```

```python
final_users.to_csv(r'json/import_files/user_data.csv', index = False)
```
